```python
# -----------------------------------------------------------------
# Decision Tree Classifier
# Predict the income of an adult based on the census data
# -----------------------------------------------------------------

# Import libraries
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  
import seaborn as sns 
from sklearn.model_selection import train_test_split 
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
import statistics as st
```

```python
# Read dataset
df=pd.read_csv('decisiontreeAdultIncome.csv')
df
```

```python
# Check for Null values
df.isnull().sum()
```

```python
# Create Dummy variables


```

```python
df=pd.get_dummies(df)
```

```python
# Create X and Y Variables
X = df.iloc[:, :-1]
X
```

```python
Y=df.iloc[:, -1:]
Y
```

```python
# Split the X and Y dataset into training and testing set
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2)
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
pred_cv=model.predict(x_test)
```

```python
# Import and train classifier
from sklearn.metrics import classification_report
classification_report(y_test,pred_cv)
```

```python
# Test the model
```

```python
# Evaluate the model
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python

ac=accuracy_score(y_test,pred_cv)
ac
```

```python
martix=confusion_matrix(y_test,pred_cv)
martix
```
